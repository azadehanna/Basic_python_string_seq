## How manage text  (string) in python : Basics

A string in Python is a sequence of characters used to represent text.

A string is anything written inside quotes(single , double  or triple).It can include letters, numbers, spaces, symbols, or even empty text.

use the = symbol to assign a string (sequence) to a variable name.


```python
dna_seq1 = 'ATGCGTACGTTA'   ## single quotes Good for short, simple sequences
print(dna_seq1)

dna_seq2 = "ATGCGTACGTTA"   ##double quotes often preferred when the string contains a single quote inside it
print(dna_seq2)

##Triple quotes used for multi-line DNA sequences or large block sequences.

dna_seq3 = """ATGCGTACGTTA       
GGCATCGTTAGC
TTAGCGATCGAT"""
print(dna_seq3)

##assign a string (sequence) to a variable name
dna_seq = "ATGCGTACGTTA"    
print("DNA:", dna_seq)
```

### String Operators in Python for Bioinformatics

The table below summarizes common Python string operators and how they are used in bioinformatics (e.g., DNA/protein sequence analysis).

### Python Operators for String Sequences (Bioinformatics Context)

| **Operator** | **Purpose**                          | **Bioinformatics Use**                                 |
|--------------|---------------------------------------|---------------------------------------------------------|
| `+`          | Concatenate strings                   | Merge exons, build gene sequences                      |
| `*`          | Repeat a string                       | Simulate repeat motifs (like CAG repeats)              |
| `in`         | Check substring (membership test)     | Check for start/stop codons, motifs                    |
| `not in`     | Check absence (membership test)       | Filter out unwanted motifs                             |
| `==`, `!=`   | Equality check (comparison)           | Mutation detection, SNP comparison                    |
| `<`, `>`, `<=`, `>=` | Lexicographic comparison (alphabetical) | Sort or rank sequences                          |
| `[]`         | Indexing                              | Access a base/amino acid                               |
| `[:]`        | Slicing                               | Extract subsequence (like codons)                     |


```python
exon1 = "ATG"
exon2 = "GGA"
gene = exon1 + exon2
print(gene)  # ATGGGA

repeat_motif = "CAG" * 3
print(repeat_motif)  # CAGCAGCAG

dna = "ATGCGTACG"
print("CGT" in dna)     # True
print("TATA" not in dna)  # True

seq1 = "ATG"
seq2 = "ATC"
print(seq1 == seq2)   # False
print(seq1 != seq2)   # True

"""What Is Lexicographic Comparison?
It's a character-by-character comparison, just like how words are sorted in a dictionary. Python compares strings using the ASCII (Unicode) values of characters.

"""

seq1 = "ATG"
seq2 = "ATA"

print(seq1 > seq2)   # True
print(seq1 < seq2)   # False

### A == A  ,   T == T  , G > A → because ASCII('G') = 71, ASCII('A') = 65

sequences = ["ATG", "AAC", "GGA", "AAA", "TGC"]
sorted_sequences = sorted(sequences)
print(sorted_sequences)    #['AAA', 'AAC', 'ATG', 'GGA', 'TGC']

## Sorted alphabetically like in a gene database or FASTA headers.
```

## Indexing means accessing individual characters in a string using their position (index).

Python uses zero-based indexing:

First character → index 0  

Second character → index 1

Last character → index -1

Seq:           A     T     G     C     G      A  
Index :        0     1     2     3     4      5  
Index (-):    -6    -5    -4    -3    -2     -1 

You cannot access an index beyond the length of the string

Index must be an integer

## Slicing means extracting a portion of a string (substring) using start:stop positions.

string[start:stop]

Includes characters from start up to but not including stop

Indexing is zero-based

You can also check for a substring using the in operator

string[start : stop : step]

start: 	Where to begin (inclusive)	      

stop: 	Where to end (exclusive)	      

step:	How many positions to skip	(1 = every base, 2 = every other)


```python
# length of sequence  (text)
seq="ATTCGATCGAT"
print(len(seq))  #11 
print(range (len(seq))) #(0,11)

#Access base in sequence  --> indexing start from 0 and from end from -1
print (seq[0]) # first base
print (seq[-1]) # last base

#Slicing (Substrings)  ---> substring = seq[start:stop]   start is inclusive, stop is exclusive.

print (seq[0:3])  # first codon  or first 3 base
print (seq[-3:])  # last 3 base
print(seq[2:4])
print(seq[::-1])   # reverse 
print(seq[:-1])   # from start but not last included
print (seq[::2]) # every other base
print(seq[len(seq)//2]) # middle base
print(seq[1:-1])    # all but not first and last
print (seq[0::3])
print (seq[1::3])
```

Padding means adding extra characters (usually spaces or zeros) to a string to make it a specific length or alignment.

It’s often used to:

Align text in tables

Format sequence data

Standardize output

Make all strings equal length for display or processing


```python
seq = "atgCGt"

print(seq.upper())      # ATGCGT — Normalize DNA
print(seq.lower())      # atgcgt — Normalize  RNA
print("bcr1".capitalize()) #Bcr1— Format gene names
print("my gene".title())# My Gene — Report formatting
print(seq.swapcase())   # ATGcgT — Case toggling (rare)

dna = "ATGCGTACGTAG"
print(dna.find("CGT"))       # 3 — First motif position (index of pattern)
print(dna.rfind("CGT"))      # 7 — Last motif position
print(dna.index("ATG"))      # 0 — Start codon
print(dna.count("G"))        # 3 — G count
print(dna.startswith("ATG")) # True — Is start codon?
print(dna.endswith("TAG"))   # True — Is stop codon?

fasta_header = "  >gene1 Homo sapiens\n"
print(fasta_header.strip())         # ">gene1 Homo sapiens" — Cleanup
print(fasta_header.lstrip())        # Remove leading whitespace
print(fasta_header.rstrip())        # Remove trailing newline
print("ATGCTT".replace("T", "U"))   # "AUGCUU" — DNA → RNA transcription

sequence = "ATG CTT CGA"
print(sequence.split())             # ['ATG', 'CTT', 'CGA'] — Codons
multiline = "ATG\nCTT\nCGA"
print(multiline.splitlines())       # ['ATG', 'CTT', 'CGA'] — FASTA blocks

print("ATG-CGA-TAA".partition("-")) # ('ATG', '-', 'CGA-TAA') — Gene region split
print("ATG-CGA-TAA".rsplit("-", 1)) # ['ATG-CGA', 'TAA'] — Last exon
print("-".join(["ATG", "CGA", "TAA"])) # "ATG-CGA-TAA" — Join codons

gene = "BRCA1"
dna = "ATGC"
padding = "   "
digits = "12345"

print(gene.isalnum())   # True — Gene name valid?
print(dna.isalpha())    # True — DNA letters only?
print(digits.isdigit()) # True — Quality scores?
print(padding.isspace())# True — Padding line in FASTA
print(gene.isupper())   # True — All uppercase?
print(gene.islower())   # False
print("Gene Name".istitle()) # True — Clean title format?

gene = "BRCA1"

print(gene.center(10))   # "  BRCA1   " — Align table
print(gene.ljust(10))    # "BRCA1     "
print(gene.rjust(10))    # "     BRCA1"
print("42".zfill(5))     # "00042" — Pad ID
print("Gene: {}".format("BRCA1"))  # "Gene: BRCA1" — Report formatting

sequence = "ATGC"
# For file writing or network transfer
encoded = sequence.encode()      # b'ATGC'
decoded = encoded.decode()       # "ATGC"
```
